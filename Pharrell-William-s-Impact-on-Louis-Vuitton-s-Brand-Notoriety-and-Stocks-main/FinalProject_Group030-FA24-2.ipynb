## Louis Vuitton Stock Data (All Time)

```python
#imported pandas library 
import pandas as pd
file_path = "LVMHF.csv"

#reading file to analyze data 
stock_data = pd.read_csv(file_path)

#creating Date column to reformat date and time for easier manipulation of data 
stock_data['Date'] = pd.to_datetime(stock_data['Date'], errors='coerce')
print(stock_data.isnull().sum()) 

#dropping rows with missing values in the date or close columns 
stock_data.dropna(subset=['Date', 'Close'], inplace=True)
stock_data.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Adj_Close', 'Volume']

#display first five rows of data frame
print(stock_data.head())
```

## ‘Pharrell Louis Vuitton’ Google Trends Data (09/01/2022-09/01/2024)

```python
import pandas as pd
file_path = "multiTimeline(Pharrell Louis Vuitton ).csv"

#reading data frame, skip first 2 rows
data = pd.read_csv(file_path, header=2)
data.columns = ['Week', 'Pharrell_LV_Interest']

#convert Week column to dataframe format to easily manipulate data 
data['Week'] = pd.to_datetime(data['Week'], errors='coerce')

#drop any rows with missing data; data cleaning 
data.dropna(subset=['Week'], inplace=True)
print(data.head())
```

```python
#Data column is in dataframe format 
stock_data['Date'] = pd.to_datetime(stock_data['Date'], errors='coerce')
data['Week'] = pd.to_datetime(data['Week'], errors='coerce')

#merged both data sets based on closest date
#merge_asof helps in aligning both dataframes based on most recent Week value 
merged_data = pd.merge_asof(
    stock_data.sort_values('Date'), 
    data.sort_values('Week'), 
    left_on='Date', 
    right_on='Week', 
    direction='backward'
)

#filtered merge based on dates listed below 
start_date = '2022-09-01'
end_date = '2023-08-01'
filtered_data = merged_data[(merged_data['Date'] >= start_date) & (merged_data['Date'] <= end_date)]
```

## ‘Louis Vuitton’ Google Trends Data (09/01/2022-09/01/2024)

```python
import pandas as pd
file_path = "LV Google Trends(LV Interest).csv"

#reading csv file and skipping first two rows 
data = pd.read_csv(file_path, header=2)

#renaming columns
data.columns = ['Week', 'LV_Interest']

#changing Week column to datetime format 
data['Week'] = pd.to_datetime(data['Week'], errors='coerce')

#drop any rows with missing data 
data.dropna(subset=['Week'], inplace=True)
print(data.head())
```

# Results

## Exploratory Data Analysis


### Louis Vuitton Stock Prices vs Search Interest

```python
import matplotlib.pyplot as plt

#create figure with size listed below 
plt.figure(figsize=(12, 6))

#plot Close and Date column, line is labeled Closing Price 
plt.plot(filtered_data['Date'], filtered_data['Close'], label='Closing Price')

#plot Pharrell_LV_Interest column against Date column, line labeled Search Interest 
plt.plot(filtered_data['Date'], filtered_data['Pharrell_LV_Interest'], label='Search Interest', alpha=0.7)

#marks Pharrell's appointment date 
plt.axvline(pd.Timestamp('2023-02-14'), color='red', linestyle='--', label='Pharrell Appointment')
plt.legend()

#title of graph listed below 
plt.title('Louis Vuitton Stock Prices vs Search Interest')
plt.show()
```

The blue line representing the closing prices of Louis Vuitton’s stock shows a consistent trend from September 2022 to July 2023, showing general growth in the company. The orange line representing search interest relating to Pharrell Williams is relatively flat or close to zero for most of the graph, with only noticeable exceptions in February and June of 2023. The red dashed line at the February 14, 2023 mark corresponds to the surge in search interest for Pharrel Williams in February 2023, which likely is due to the attention his announcement as the Men’s Creative got. The spike in search interest in June is due to the fact that Pharrell's collection debuted in June.Regarding the relationship between stock prices and search interest, there is not enough of a direct impact seen in this graph to make a concrete conclusion that they are correlated, since the stock prices were already increasing. 

### Monthly Average Closing Price

```python
filtered_data['Month'] = filtered_data['Date'].dt.to_period('M')

#groups data by Month column and calculates average closing price for each month 
monthly_avg = filtered_data.groupby('Month')['Close'].mean().reset_index()

#create bar chart for monthly average closing price 
plt.bar(monthly_avg['Month'].astype(str), monthly_avg['Close'])
plt.xticks(rotation=45)

#set title of graph 
plt.title('Monthly Average Closing Price')
plt.show()
```

 The monthly average closing price, as shown by the bar plots, from September 2022 to April 2023 shows an overall increase in Louis Vuitton’s stock value. It dips after April until July 2023 when it spikes up again. From July to August Louis Vuitton’s stock price reduces. This depicts a general trend of LV’s general closing price.

### Pharrell Louis Vuitton Search Interest vs Appointment Date

```python
plt.figure(figsize=(12, 6))

#plots weekly Pharrell LV search interest data, Week column and Pharrell_LV_Interest columns  
plt.plot(filtered_data['Week'], filtered_data['Pharrell_LV_Interest'], label='Pharrell LV Search Interest', color='blue')

#Pharell’s appointment date
announcement_date = pd.Timestamp('2023-02-14')
plt.axvline(announcement_date, color='red', linestyle='--', label='Pharrell Appointment (2023-02-14)')

plt.legend()

#title for graph set 
plt.title('Pharrell Louis Vuitton Search Interest vs Appointment Date')
plt.xlabel('Date')
plt.ylabel('Search Interest')
plt.xticks(rotation=45)

plt.show()
```

The graph representing search interest relating to Pharrell Williams is relatively flat or close to zero for most of the graph, with only noticeable exceptions in February and June of 2023. The red dashed line at the February 14, 2023 mark corresponds to the surge in search interest for Pharrel Williams in February 2023, which likely is due to the attention his announcement as the Men’s Creative got.

### Correlation Coefficient Heatmap

```python
import seaborn as sns
#calculates correlation matrix for Close and Pharell_LV_Interest columns 
correlation_matrix = merged_data[['Close', 'Pharrell_LV_Interest']].corr()

plt.figure(figsize=(6, 4))

#generates heat map for correlation matrix, displays values on map and sets colors for visual representation of heat map
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

#title set for visual 
plt.title('Correlation Coefficient Heatmap: Louis Vuitton Stock vs Pharrell Interest')
plt.show()
```
